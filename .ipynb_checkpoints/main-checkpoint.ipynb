```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
V = np.arange(2,41,2)
V
```

```python
T = np.array([77, 77, 77.5, 77.5, 78, 78.5, 78.5, 80.5, 84.5, 85.5, 
              86.5, 87.5, 88, 134, 137, 137.5, 137.7, 138, 138, 138.1])
T
```

```python
df = pd.DataFrame({"Объем, мл": V, 
              "Температура, °C": T})
```

```python
fig, ax = plt.subplots(1,1)
df.plot(x="Объем, мл", y="Температура, °C", ax=ax, style=".-", title="Кривая ректификации", grid=True)
ax.plot(np.linspace(0,50,10),[T[T<85].mean()] * 10, label="$T_{кип}$ этанола")
ax.plot(np.linspace(0,50,10),[T[T>100].mean()] * 10, label="$T_{кип}$ амилового спирта")
plt.ylabel("Температура, °C")
plt.legend()
plt.show()
```

**Данные, полученные из графика:**  


$T_{кип}(C_2H_5OH) \approx 78.8 °C$   
$T_{кип}(C_5H_{11}OH) \approx 137.2 °C$

**Табличные данные:**  
Показатель преломления $C_2H_5OH$ = 1.36  
Показатель преломления $C_5H_{11}OH$ = 1.41

**Экспериментально полученные данные:**  
Показатель преломления фракции №1 = 1.372  
Показатель преломления фракции №2 = 1.407

Отсюда получаем:

Фракция №1:  $ \Phi_{C_2H_5OH} = 0.76 $  
            $ \Phi_{C_5H_{11}OH} = 0.24 $

```python
df1 = pd.DataFrame(data=np.array([[0.76, 0.24], [0.06, 0.94]]),index=["Фракция №1", "Фракция №2"], 
                   columns =[["Объемная доля", "Объемная доля"],["C2H5OH","C5H11OH"]])
```

```python
df1
```

```python
df1.plot.barh(title="Состав полученных фракций",stacked=True)
plt.xlabel("Объемная доля")
plt.xlim(0,1)
plt.legend()
plt.show()
```
